# 手动反向传播

## Karpathy 设置偏置不为零的原因

Karpathy 在这里的思路其实是为了调试便利性和避免隐藏潜在错误：
	•	如果偏置（b）初始化为全零，虽然在理论上是合理的（因为偏置不会影响梯度传播的对称性），但这种“完美对称”有时会掩盖代码中的梯度计算错误。
	•	当所有初始值为零时，很多梯度表达式会被“简化”，导致某些实现错误在训练早期并不明显暴露出来。
	•	所以，Karpathy故意将偏置设置为小的随机数，这样可以打破这种“零简化”的现象，让梯度流更接近真实训练中的表现，方便在调试时发现问题。

总结一句话：


偏置设置为小随机数，并不是为了性能，而是为了在开发和调试阶段，更容易暴露梯度计算或参数更新中的潜在 bug。
